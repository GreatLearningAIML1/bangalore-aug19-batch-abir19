```python
import re, string

def clean_str(string):
  """String cleaning before vectorization
  """
  
  try:    
    string = re.sub(r'^https?:\/\/<>.*[\r\n]*', '', string, flags=re.MULTILINE)
    string  = re.sub(r"[^A-Za-z]", " ", string)         
    words =string.strip().lower().split()    
    words = [w for w in words if len(w)>=1]
    return " ".join(words)	
  except:
    return ""
```

```python
blogText_df["clean_text"] = blogText_df["text"].apply(clean_str)
```

```python
blogText_df.pop("text")
```

```python
blogText_df.dtypes
```

```python
blogText_df['age'] = blogText_df['age'].astype(str)
blogText_df['gender'] = blogText_df['gender'].astype(str)
```

```python
blogText_df['labels'] = blogText_df[['gender','age','topic','sign']].agg(','.join,axis=1)
```

```python
blogText_df
```

```python
blogText_df.drop(['id','gender','age','topic','sign','date'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test, y_train, y_test = train_test_split(blogText_df.clean_text, blogText_df.labels, random_state=30)
```

```python
#Traing data
print(df_train.shape)
print(df_test.shape)
```

```python
# import and instantiate CountVectorizer (with the default parameters)
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cvect = CountVectorizer(stop_words='english',ngram_range=(1,2))
```

```python
cvect.fit(df_train)
```

```python
#Check the vocablury size
len(cvect.vocabulary_)
```

```python
X_train = cvect.transform(df_train)
X_test = cvect.transform(df_test)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
labelVector = CountVectorizer()
label_ct =labelVector.fit(blogText_df['labels'])
dict_ =label_ct.vocabulary_
```

```python
dict_
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
Y_train_mlb= mlb.fit_transform(y_train)
```

```python
Y_test_mlb = mlb.transform(y_test)
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(solver='lbfgs')
clf = OneVsRestClassifier(clf)
```

```python
clf.fit(X_train,Y_train_mlb)
```

```python
from sklearn import metrics
metrics.accuracy_score(Y_test_mlb,clf.predict(X_test))
```

```python
from sklearn.metrics import classification_report
y_predict = clf.predict(X_test)
print(classification_report(Y_test_mlb,y_predict))
```
